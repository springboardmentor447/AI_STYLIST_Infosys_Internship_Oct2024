```python
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.preprocessing.image import load_img, img_to_array
import numpy as np
import os

# Load the pre-trained VGG16 model without the top layer
vgg16_model = VGG16(weights='imagenet', include_top=False)

# Function to extract embeddings for an image
def extract_embeddings(image_path):
    try:
        # Load the image
        img = load_img(image_path, target_size=(224, 224))  # VGG16 requires 224x224 images
        img_array = img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = preprocess_input(img_array)

        # Extract features
        embeddings = vgg16_model.predict(img_array)
        return embeddings.flatten()
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None

# Path to the image directory
image_dir = r"C:\Users\ANGELIN\Downloads\images"

# Generate embeddings for all images in the directory
image_embeddings = {}
for image_file in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_file)
    image_embeddings[image_file] = extract_embeddings(image_path)

# Save embeddings as a CSV file
embeddings_df = pd.DataFrame.from_dict(image_embeddings, orient='index')
embeddings_df.to_csv('image_embeddings.csv', index=True)
print("Image embeddings saved successfully!")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to recommend similar images
def recommend_images(image_name, top_n=5):
    if image_name not in image_embeddings:
        print("Image not found in the dataset.")
        return

    target_embedding = np.array(image_embeddings[image_name]).reshape(1, -1)
    similarities = {}

    # Calculate cosine similarity
    for img, embedding in image_embeddings.items():
        if img != image_name:  # Exclude the target image itself
            similarity = cosine_similarity(target_embedding, np.array(embedding).reshape(1, -1))
            similarities[img] = similarity[0][0]

    # Get top N similar images
    sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]
    print(f"\nTop {top_n} recommendations for '{image_name}':")
    for img, score in sorted_similarities:
        print(f"{img}: Similarity Score = {score:.4f}")

# Test the recommendation function
test_image = "sample_image.jpg"  # Replace with an actual image file name
recommend_images(test_image, top_n=5)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Function to display images
def display_recommendations(image_name, top_n=5):
    if image_name not in image_embeddings:
        print("Image not found in the dataset.")
        return

    target_embedding = np.array(image_embeddings[image_name]).reshape(1, -1)
    similarities = {}

    # Calculate cosine similarity
    for img, embedding in image_embeddings.items():
        if img != image_name:
            similarity = cosine_similarity(target_embedding, np.array(embedding).reshape(1, -1))
            similarities[img] = similarity[0][0]

    # Get top N similar images
    sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]

    # Plot the target image
    fig, axes = plt.subplots(1, top_n + 1, figsize=(15, 5))
    axes[0].imshow(Image.open(os.path.join(image_dir, image_name)))
    axes[0].set_title("Target Image")
    axes[0].axis('off')

    # Plot the recommended images
    for i, (img, _) in enumerate(sorted_similarities):
        img_path = os.path.join(image_dir, img)
        axes[i + 1].imshow(Image.open(img_path))
        axes[i + 1].set_title(f"Recommendation {i + 1}")
        axes[i + 1].axis('off')

    plt.show()

# Test the display function
display_recommendations(test_image, top_n=5)
```
