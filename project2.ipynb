```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tqdm import tqdm

# Set up the path and image dimensions
path = "C:/Users/sjvar/Downloads/extracted/"
img_width, img_height = 224, 224

# Load the pre-trained MobileNetV2 model (without the top layer, for feature extraction)
model1 = MobileNetV2(weights="imagenet", include_top=False, pooling="avg")

# Function to get image path
def img_path(img):
    return path + 'images/' + img

# Function to predict and extract embeddings from the model
def model_predict(model, img_name):
    # Load the image and resize it
    img = image.load_img(img_path(img_name), target_size=(img_width, img_height))
    
    # Convert image to array
    x = image.img_to_array(img)
    
    # Expand the dimensions to match model input
    x = np.expand_dims(x, axis=0)
    
    # Preprocess the input (similar to what is done in the original model)
    x = preprocess_input(x)
    
    # Generate the prediction (embedding)
    return model.predict(x).reshape(-1)

# Read the CSV file into a pandas DataFrame
csv_file = r"C:/Users/sjvar/Downloads/extracted/styles_with_images.csv"
df = pd.read_csv(csv_file)

# Limit the processing to 5,000 images
batch_size = 5000
total_images = min(len(df), batch_size)

# Get the subset of the DataFrame for the first 5,000 images
batch_df = df.iloc[:total_images]

# Initialize the list for embeddings
embeddings = []

# Use tqdm to track progress for the current batch
for img_name in tqdm(batch_df['image'], desc="Processing 5,000 images", unit="image"):
    try:
        # Get the embedding for each image
        embedding = model_predict(model1, img_name)
        embeddings.append(embedding)  # Append the embedding to the list
    except Exception as e:
        print(f"Error processing image {img_name}: {e}")

# Convert the embeddings list into a DataFrame
embeddings_df = pd.DataFrame(embeddings)

# Merge embeddings with the current batch DataFrame
batch_embedding_df = pd.concat([batch_df.reset_index(drop=True), embeddings_df], axis=1)

# Print the first 5 rows of the DataFrame with embeddings
print(batch_embedding_df.head())  # Print the first 5 rows

# Save the DataFrame with embeddings to a CSV file
batch_embedding_df.to_csv(r"C:/Users/sjvar/Downloads/extracted/styles_with_embeddings_5000.csv", index=False)

print(f"Completed generating embeddings for {len(batch_embedding_df)} images.")
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"C:\Users\sjvar\Downloads\extracted\styles_with_embeddings_5000.csv"  # Replace with your actual path
df = pd.read_csv(file_path)

# Limit to the first 5,000 rows
df = df.head(5000)

# Specify the embedding columns (assumes columns are labeled numerically for embeddings)
embedding_columns = [col for col in df.columns if col.isdigit()]

# Check if embeddings are present
if not embedding_columns:
    raise ValueError("No embedding columns found in the dataset!")

# Function to recommend similar items
def recommend_similar_items(item_id, num_recommendations=5):
    # Check if item_id exists in the first 5,000 rows
    if item_id not in df['id'].values:
        raise ValueError(f"Item with id {item_id} not found in the first 5,000 items of the dataset!")

    # Get the embedding for the given item
    item_embedding = df[df['id'] == item_id][embedding_columns].values

    # Calculate cosine similarity with all items
    embeddings_matrix = df[embedding_columns].values
    similarities = cosine_similarity(item_embedding, embeddings_matrix)[0]

    # Get indices of the most similar items (excluding the item itself)
    similar_indices = similarities.argsort()[-(num_recommendations + 1):][::-1]
    similar_indices = [idx for idx in similar_indices if df.iloc[idx]['id'] != item_id]

    # Fetch metadata and images for the recommendations
    recommendations = df.iloc[similar_indices]
    return recommendations

# Function to display recommendations with images
def display_recommendations(item_id, recommendations):
    # Get the query item's metadata
    query_item = df[df['id'] == item_id].iloc[0]
    print(f"Query Item: {query_item['productDisplayName']}")
    
    # Display the query image
    plt.figure(figsize=(5, 5))
    query_image_path = r"C:\Users\sjvar\Downloads\extracted\images"  # Replace with your actual image directory
    query_image = Image.open(f"{query_image_path}/{query_item['image']}")
    plt.imshow(query_image)
    plt.title("Query Image")
    plt.axis("off")
    plt.show()

    # Display recommendation images
    for _, row in recommendations.iterrows():
        print(f"Recommended Item: {row['productDisplayName']}")
        try:
            plt.figure(figsize=(5, 5))
            image_path = f"{query_image_path}/{row['image']}"
            recommended_image = Image.open(image_path)
            plt.imshow(recommended_image)
            plt.title(row['productDisplayName'])
            plt.axis("off")
            plt.show()
        except FileNotFoundError:
            print(f"Image not found for: {row['productDisplayName']}")

# Example Usage
try:
    item_id = 15970  # Replace with the ID of an item in your dataset
    num_recommendations = 5
    recommendations = recommend_similar_items(item_id, num_recommendations)
    display_recommendations(item_id, recommendations)
except Exception as e:
    print(f"Error: {e}")
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import torch
from torchvision import models, transforms
from tqdm import tqdm

# Load the dataset
file_path = r"C:\Users\sjvar\Downloads\extracted\styles_with_embeddings_5000.csv"  # Replace with your actual path
df = pd.read_csv(file_path)

# Limit to the first 1,000 rows
df = df.head(1000)

# Load pre-trained DenseNet model
model = models.densenet121(pretrained=True)
model = torch.nn.Sequential(*(list(model.children())[:-1]))  # Remove classification layer
model.eval()

# Transformation for DenseNet
image_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Directory path for images
query_image_path = r"C:\Users\sjvar\Downloads\extracted\images"  # Replace with your actual image directory

# Function to extract DenseNet embeddings
def extract_densenet_embedding(image_path):
    try:
        image = Image.open(image_path).convert('RGB')  # Ensure the image is RGB
        image = image_transform(image).unsqueeze(0)  # Add batch dimension
        with torch.no_grad():
            embedding = model(image).squeeze().numpy()  # Remove batch dimension
        return embedding
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None

# Extract embeddings for all images
def generate_embeddings(df):
    embeddings = []
    for _, row in tqdm(df.iterrows(), total=len(df)):
        image_path = f"{query_image_path}/{row['image']}"
        embedding = extract_densenet_embedding(image_path)
        if embedding is not None:
            embeddings.append(embedding)
        else:
            embeddings.append([0] * 1024)  # DenseNet default output size is 1024
    return np.array(embeddings)

# Generate embeddings (once for all items)
print("Generating embeddings for 1,000 images...")
df_embeddings = generate_embeddings(df)

# Function to recommend similar items
def recommend_similar_items(item_id, num_recommendations=5):
    # Check if item_id exists
    if item_id not in df['id'].values:
        raise ValueError(f"Item with id {item_id} not found in the dataset!")

    # Get the embedding for the given item
    item_index = df[df['id'] == item_id].index[0]
    item_embedding = df_embeddings[item_index].reshape(1, -1)

    # Calculate cosine similarity with all items
    similarities = cosine_similarity(item_embedding, df_embeddings)[0]

    # Get indices of the most similar items (excluding the item itself)
    similar_indices = similarities.argsort()[-(num_recommendations + 1):][::-1]
    similar_indices = [idx for idx in similar_indices if df.iloc[idx]['id'] != item_id]

    # Fetch metadata and images for the recommendations
    recommendations = df.iloc[similar_indices]
    return recommendations

# Function to display recommendations with images
def display_recommendations(item_id, recommendations):
    # Get the query item's metadata
    query_item = df[df['id'] == item_id].iloc[0]
    print(f"Query Item: {query_item['productDisplayName']}")
    
    # Display the query image
    plt.figure(figsize=(5, 5))
    query_image = Image.open(f"{query_image_path}/{query_item['image']}")
    plt.imshow(query_image)
    plt.title("Query Image")
    plt.axis("off")
    plt.show()

    # Display recommendation images
    for _, row in recommendations.iterrows():
        print(f"Recommended Item: {row['productDisplayName']}")
        try:
            plt.figure(figsize=(5, 5))
            image_path = f"{query_image_path}/{row['image']}"
            recommended_image = Image.open(image_path)
            plt.imshow(recommended_image)
            plt.title(row['productDisplayName'])
            plt.axis("off")
            plt.show()
        except FileNotFoundError:
            print(f"Image not found for: {row['productDisplayName']}")

# Example Usage
try:
    item_id = 15970  # Replace with the ID of an item in your dataset
    num_recommendations = 5
    recommendations = recommend_similar_items(item_id, num_recommendations)
    display_recommendations(item_id, recommendations)
except Exception as e:
    print(f"Error: {e}")
```
