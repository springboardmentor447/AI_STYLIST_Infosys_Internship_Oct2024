```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px  # For interactive visualizations
import plotly.graph_objects as go
```

```python
# Load the LDJSON file
data = pd.read_json('C:/Users/ADMIN/Downloads/fashion_products_data.ldjson', lines=True)

# Quick look at the dataset
print(data.head())
```

```python
# Check for missing values and fill with mean/median if necessary
# Fill only the numeric columns with their mean values
data.fillna(data.mean(numeric_only=True), inplace=True)


# Optional: Drop columns or rows with too many missing values
data.dropna(thresh=0.9 * data.shape[1], axis=0, inplace=True)  # Drops rows with 90% missing values
```

```python
# Assuming there’s a 'month' column, we can use a barplot
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='month', palette='viridis')
plt.title("Distribution of Purchases by Month")
plt.xlabel("Month")
plt.ylabel("Number of Purchases")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure 'crawl_timestamp' is in datetime format
data['crawl_timestamp'] = pd.to_datetime(data['crawl_timestamp'], errors='coerce')

# Extract the month from 'crawl_timestamp' and create a new column 'month'
data['month'] = data['crawl_timestamp'].dt.month

# Plot the distribution of purchases by month
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='month', palette='viridis')
plt.title("Distribution of Purchases by Month")
plt.xlabel("Month")
plt.ylabel("Number of Purchases")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(data=data, y='sales_price', x='browsenode', palette='coolwarm')
plt.title("Sales Price Distribution Across Different Categories")
plt.xticks(rotation=45)
plt.ylabel("Sales Price")
plt.xlabel("Category")
plt.show()
```

```python
# Extracting month from crawl_timestamp
data['crawl_timestamp'] = pd.to_datetime(data['crawl_timestamp'])
data['month'] = data['crawl_timestamp'].dt.month_name()  # Get month names

plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='month', palette='viridis', order=data['month'].value_counts().index)
plt.title("Distribution of Purchases by Month")
plt.xlabel("Month")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, y='sales_price', x='brand', palette='Set2')
plt.title("Sales Price Distribution by Brand")
plt.xticks(rotation=45)
plt.ylabel("Sales Price")
plt.xlabel("Brand")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='delivery_type', palette='pastel')
plt.title("Count of Products by Delivery Type")
plt.xlabel("Delivery Type")
plt.ylabel("Number of Products")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='rating', y='sales_price', hue='delivery_type', palette='bright', alpha=0.7)
plt.title("Sales Price vs. Rating")
plt.xlabel("Rating")
plt.ylabel("Sales Price")
plt.legend(title='Delivery Type')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```
