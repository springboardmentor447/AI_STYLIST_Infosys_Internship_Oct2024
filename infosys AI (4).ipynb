```python
import nltk
from nltk.corpus import stopwords

# Download stopwords from NLTK
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Example text data
text_data = "This is an example of text processing using NLTK in Python."

# Tokenize text and remove stop words
words = text_data.split()  # Tokenize by splitting on whitespace
filtered_words = [word for word in words if word.lower() not in stop_words]

print("Original text:", text_data)
print("Text without stop words:", " ".join(filtered_words))
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Download NLTK resources
nltk.download('punkt')

text = "Natural Language Processing with NLTK is quite interesting. Let's explore some basics!"

# Sentence Tokenization
sentences = sent_tokenize(text)
print("Sentences:", sentences)

# Word Tokenization
words = word_tokenize(text)
print("Words:", words)
```

```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()

# Sample words
words = ["running", "runs", "easily", "fairly"]

# Stem each word
stemmed_words = [stemmer.stem(word) for word in words]
print("Stemmed words:", stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('wordnet')
nltk.download('omw-1.4')  # For additional language support

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize words with POS context
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print("Lemmatized words:", lemmatized_words)
```

```python
nltk.download('averaged_perceptron_tagger')

# Tag parts of speech
pos_tags = nltk.pos_tag(words)
print("POS tags:", pos_tags)
```
