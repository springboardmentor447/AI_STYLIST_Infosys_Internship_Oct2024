```python
# Define a function to detect and remove outliers based on IQR
def remove_outliers_iqr(df, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    # Calculate IQR
    IQR = Q3 - Q1
    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Filter out the outliers
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

# Apply the function to relevant columns in the dataset
# For example, columns like 'sales_price' and 'rating' might have outliers
if 'sales_price' in data.columns and pd.api.types.is_numeric_dtype(data['sales_price']):
    data = remove_outliers_iqr(data, 'sales_price')

if 'rating' in data.columns and pd.api.types.is_numeric_dtype(data['rating']):
    data = remove_outliers_iqr(data, 'rating')
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from langdetect import detect, LangDetectException

# Ensure required NLTK data is downloaded
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Columns to retain
required_columns = ['asin', 'product_url', 'product_name', 'sales_price', 'rating', 'meta_keywords', 'medium', 'brand']

# Load dataset
file_path = "C:/Users/varsh/Downloads/fashion_products_data.ldjson"
data = pd.read_json(file_path, lines=True)

# Retain only the required columns
data = data[required_columns]
```

```python
def remove_non_english_words(text):
    words = text.split()
    cleaned_words = []
    for word in words:
        try:
            if detect(word) == 'en':  # Check if word is in English
                cleaned_words.append(word)
        except LangDetectException:
            pass  # Ignore words that cannot be detected
    return ' '.join(cleaned_words)

# Apply the function to the 'product_name' column
data['product_name'] = data['product_name'].apply(remove_non_english_words)
```

```python
# Dictionary for expanding abbreviations
abbreviations = {
    "nxt": "next",
    "btw": "by the way",
    "u": "you",
    "ur": "your"
}

def expand_short_forms(text):
    words = text.split()
    expanded_words = [abbreviations.get(word, word) for word in words]
    return ' '.join(expanded_words)

# Apply the function to the 'product_name' column
data['product_name'] = data['product_name'].apply(expand_short_forms)
```

```python
def apply_stemming(text):
    words = text.split()
    stemmed_words = [stemmer.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Apply stemming to the 'product_name' column
data['stemmed_product_name'] = data['product_name'].apply(apply_stemming)
```

```python
def apply_lemmatization(text):
    words = text.split()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)

# Apply lemmatization to the 'stemmed_product_name' column
data['lemmatized_product_name'] = data['stemmed_product_name'].apply(apply_lemmatization)
```

```python
# Save the final cleaned dataset to a new file
output_path = "C:/Users/varsh/Downloads/cleaned_fashion_products_data.csv"
data.to_csv(output_path, index=False)

# Print a preview of the cleaned data
print(data.head())
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

# Ensure required NLTK data is downloaded
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize stopwords and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Update your cleaning function
def clean_product_name(text):
    # Step 1: Remove special characters and lowercase the text
    text = re.sub(r'[^a-zA-Z\s]', '', text).lower()

    # Step 2: Tokenize the text
    words = text.split()

    # Step 3: Remove stopwords and filter out short words
    words = [word for word in words if word not in stop_words and len(word) > 2]

    # Step 4: Perform lemmatization
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

    return ' '.join(lemmatized_words)

# Load your data
file_path = "C:/Users/varsh/Downloads/fashion_products_data.ldjson"
data = pd.read_json(file_path, lines=True)

# Apply the cleaning function
data['lemmatized_product_name'] = data['product_name'].apply(clean_product_name)

# View the output
print(data[['asin', 'product_name', 'lemmatized_product_name']].head())

# Optional: Save the cleaned data to a new file
output_file = "C:/Users/varsh/Downloads/cleaned_fashion_products.csv"
data[['asin', 'product_name', 'lemmatized_product_name']].to_csv(output_file, index=False)
print(f"Cleaned data saved to {output_file}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load dataset
file_path = "C:/Users/varsh/Downloads/cleaned_fashion_products.csv"
data = pd.read_csv(file_path)

# Fill missing values in 'lemmatized_product_name' column
data['lemmatized_product_name'] = data['lemmatized_product_name'].fillna('')

# Vectorizing the product names
vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(data['lemmatized_product_name'])

# Recommendation function
def recommend_products(product_id, num_products, vectors):
    # Find the index of the given product_id
    if product_id not in data['asin'].values:
        raise ValueError(f"Product ID {product_id} not found in the dataset!")
    product_idx = data[data['asin'] == product_id].index[0]
    
    # Compute cosine similarity
    similarity_matrix = cosine_similarity(vectors)
    
    # Get similarity scores for the given product
    similarity_scores = list(enumerate(similarity_matrix[product_idx]))
    
    # Sort by similarity scores (descending) and exclude the product itself
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    sorted_scores = [item for item in sorted_scores if item[0] != product_idx]
    
    # Get the top N recommendations
    top_products = sorted_scores[:num_products]
    
    # Retrieve product details
    recommendations = data.iloc[[idx for idx, _ in top_products]]
    return recommendations[['asin', 'product_name']]

# Example usage
product_id = 'B07STS2W9T'  # Replace with a valid product ID
num_products = 3  # Number of products to recommend
recommended_products = recommend_products(product_id, num_products, vectors)
print(recommended_products)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import Image, display

# Load dataset
file_path = "C:/Users/varsh/Downloads/cleaned_fashion_products.csv"
data = pd.read_csv(file_path)

# Map product IDs to image paths
image_paths = {
    'B07STS2W9T': 'C:/Users/varsh/Downloads/image1.jpg',
    'B07W5X5BN4': 'C:/Users/varsh/Downloads/image2.jpg',
    'B078YRTS7B': 'C:/Users/varsh/Downloads/image3.jpg',
    # Add more product IDs and image paths as necessary
}

# Add 'image_path' column to the dataset using the mapping
data['image_path'] = data['asin'].map(image_paths)

# Fill missing values in 'lemmatized_product_name' column
data['lemmatized_product_name'] = data['lemmatized_product_name'].fillna('')

# Vectorizing the product names
vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(data['lemmatized_product_name'])

# Recommendation function
def recommend_products(product_id, num_products, vectors):
    # Find the index of the given product_id
    if product_id not in data['asin'].values:
        raise ValueError(f"Product ID {product_id} not found in the dataset!")
    product_idx = data[data['asin'] == product_id].index[0]
    
    # Compute cosine similarity
    similarity_matrix = cosine_similarity(vectors)
    
    # Get similarity scores for the given product
    similarity_scores = list(enumerate(similarity_matrix[product_idx]))
    
    # Sort by similarity scores (descending) and exclude the product itself
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    sorted_scores = [item for item in sorted_scores if item[0] != product_idx]
    
    # Get the top N recommendations
    top_products = sorted_scores[:num_products]
    
    # Retrieve product details (including 'asin', 'product_name', and 'image_path')
    recommendations = data.iloc[[idx for idx, _ in top_products]]
    
    # Use .loc to fill missing 'image_path' values
    recommendations.loc[:, 'image_path'] = recommendations['image_path'].fillna('No image available')
    
    return recommendations[['asin', 'product_name', 'image_path']]

# Example usage
product_id = 'B07STS2W9T'  # Replace with a valid product ID
num_products = 3  # Number of products to recommend
recommended_products = recommend_products(product_id, num_products, vectors)

# Display recommended products with images
for index, row in recommended_products.iterrows():
    print(f"Product Name: {row['product_name']}")
    print(f"Product ID: {row['asin']}")
    if row['image_path'] != 'No image available':
        display(Image(filename=row['image_path']))  # Display the image
    else:
        print("No image available")
    print("\n")
```
