```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset (assuming you've already loaded it as 'data')
# data = pd.read_json("C:/Users/sjvar/Downloads/fashion_products_data.ldjson", lines=True)

# Univariate analysis function for numerical columns
def univariate_numerical(data, column):
    # Convert the column to numeric, setting errors='coerce' to turn non-numeric values into NaN
    data[column] = pd.to_numeric(data[column], errors='coerce')
    
    plt.figure(figsize=(10, 5))
    # Plot histogram, dropping NaN values
    data[column].dropna().plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    
    # Print summary statistics
    print(f"Summary statistics for {column}:\n", data[column].describe(), "\n")

# Univariate analysis function for categorical columns
def univariate_categorical(data, column, top_n=10):
    plt.figure(figsize=(10, 5))
    # Get the top N most frequent categories
    top_categories = data[column].value_counts().head(top_n)
    # Plot bar chart
    top_categories.plot(kind='bar', color='coral', edgecolor='black')
    plt.title(f'Top {top_n} {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    # Print value counts
    print(f"Value counts for {column}:\n", top_categories, "\n")

# Perform univariate analysis on numerical columns
univariate_numerical(data, 'sales_price')
univariate_numerical(data, 'rating')
univariate_numerical(data, 'discount_percentage')

# Perform univariate analysis on categorical columns
univariate_categorical(data, 'brand')
univariate_categorical(data, 'colour')
univariate_categorical(data, 'amazon_prime__y_or_n')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Multivariate analysis function with conditional palette usage
def multivariate_analysis(data, plot_type, x, y=None, hue=None, top_n=None):
    plt.figure(figsize=(10, 6))
    
    if top_n:
        # Filter to top categories by frequency if 'top_n' is specified
        if data[x].nunique() > top_n:
            top_categories = data[x].value_counts().index[:top_n]
            data = data[data[x].isin(top_categories)]
    
    if plot_type == "scatter":
        sns.scatterplot(data=data, x=x, y=y, hue=hue, palette="viridis" if hue else None)
        plt.title(f'Scatter Plot of {y} vs {x}')
    
    elif plot_type == "box":
        if hue:
            sns.boxplot(data=data, x=x, y=y, hue=hue, palette="Set2")
        else:
            sns.boxplot(data=data, x=x, y=y, color="skyblue")
        plt.title(f'Box Plot of {y} by {x}')
    
    elif plot_type == "heatmap":
        plt.figure(figsize=(12, 8))
        correlation = data.corr()
        sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=0.5)
        plt.title("Correlation Heatmap")
    
    elif plot_type == "bar":
        grouped_data = data.groupby([x, hue])[y].mean().unstack()
        grouped_data.plot(kind='bar', stacked=False, color=['#5B84B1', '#FC766A'])
        plt.title(f'{y} by {x} and {hue}')
        plt.xlabel(x)
        plt.ylabel(y)

    plt.show()

# Example of using the function with limited data
# Box Plot: Limited to top 10 'brands' by count
multivariate_analysis(data, plot_type="box", x="brand", y="sales_price", top_n=10)
```

```python
import numpy as np

def remove_outliers(data, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the data to remove outliers
    filtered_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return filtered_data

# Remove outliers in the 'sales_price' column
filtered_data = remove_outliers(data, 'sales_price')

# Plotting the box plot without outliers
multivariate_analysis(filtered_data, plot_type="box", x="brand", y="sales_price", top_n=10)
```

```python
import pandas as pd
import re

# Load the data into a DataFrame (update the path as needed)
data = pd.read_json("C:/Users/sjvar/Downloads/fashion_products_data.ldjson", lines=True)

# Dictionary of common long forms and their contractions
contractions = {
    "do not": "don't",
    "cannot": "can't",
    "is not": "isn't",
    "are not": "aren't",
    "was not": "wasn't",
    "were not": "weren't",
    "will not": "won't",
    "would not": "wouldn't",
    "have not": "haven't",
    "has not": "hasn't",
    "had not": "hadn't",
    "could not": "couldn't",
    "should not": "shouldn't",
    "must not": "mustn't",
    "does not": "doesn't",
    "did not": "didn't",
    "it is": "it's",
    "there is": "there's",
    "there are": "there're",
}

# Function to replace long forms with contractions
def convert_to_contractions(text):
    # Use regex to replace each long form with its contraction
    for long_form, contraction in contractions.items():
        text = re.sub(r'\b' + re.escape(long_form) + r'\b', contraction, text, flags=re.IGNORECASE)
    return text

# Apply the function to the 'product_name' column
data['product_name'] = data['product_name'].apply(convert_to_contractions)

# Display the modified DataFrame
print(data)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
data = pd.read_json("C:/Users/sjvar/Downloads/fashion_products_data.ldjson", lines=True)

# Convert non-string columns to strings and handle NaN values
data['product_name'] = data['product_name'].fillna('').apply(str)
data['meta_keywords'] = data['meta_keywords'].fillna('').apply(str)
data['product_details__k_v_pairs'] = data['product_details__k_v_pairs'].fillna('').apply(str)

# Combine text features
data['text_features'] = (
    data['product_name'] + ' ' +
    data['meta_keywords'] + ' ' +
    data['product_details__k_v_pairs']
)

# Initialize CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_features=5000)
vectorized_data = vectorizer.fit_transform(data['text_features'])

# Compute cosine similarity
cosine_sim = cosine_similarity(vectorized_data)

# Define a recommendation function
def recommend_products(product_index, top_n=5):
    """
    Recommends top N similar products based on cosine similarity.
    """
    sim_scores = list(enumerate(cosine_sim[product_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_indices = [i[0] for i in sim_scores[1:top_n + 1]]
    return data.iloc[top_indices][['uniq_id', 'product_name', 'brand', 'sales_price']]

# Example usage
recommendations = recommend_products(product_index=12, top_n=5)
print(recommendations)
```

```python
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Step 1: Load the dataset
data_path = "C:/Users/sjvar/Downloads/fashion_products_data.ldjson"
data = pd.read_json(data_path, lines=True)

# Step 2: Check the dataset structure and ensure 'product_name' exists
if 'product_name' not in data.columns:
    raise ValueError("The column 'product_name' does not exist in the dataset. Please check the data structure.")

# Step 3: Prepare Product Names for Word2Vec
product_names = data['product_name'].fillna('').str.split()  # Tokenize product names

# Step 4: Train Word2Vec Model
word2vec_model = Word2Vec(
    sentences=product_names,  # Tokenized product names
    vector_size=100,          # Size of word embeddings
    window=5,                 # Context window size
    min_count=1,              # Minimum word frequency
    workers=4,                # Number of CPU threads
    seed=42                   # Random seed for reproducibility
)

# Step 5: Create Product Vectors by Averaging Word Vectors
def get_product_vector(words, model):
    """Generate an averaged vector for a product name based on its words."""
    vectors = [model.wv[word] for word in words if word in model.wv]
    if vectors:
        return np.mean(vectors, axis=0)  # Average the vectors
    else:
        return np.zeros(model.vector_size)  # Return a zero vector if no valid words

product_vectors = np.array([
    get_product_vector(words, word2vec_model) for words in product_names
])

# Step 6: Compute Similarity Matrix
cosine_sim = cosine_similarity(product_vectors)

# Step 7: Recommendation Function
def recommend_products(product_index, top_n=5):
    """
    Recommend products based on cosine similarity of product vectors.
    
    Args:
        product_index (int): The index of the product for which recommendations are to be made.
        top_n (int): Number of recommendations to return (default: 5).
    """
    if product_index >= len(data):
        print(f"Error: product_index {product_index} is out of range.")
        return

    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[product_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of top N similar products (excluding itself)
    top_indices = [i[0] for i in sim_scores[1:top_n + 1]]

    # Display Recommendations
    recommendations = data.iloc[top_indices][['uniq_id', 'product_name', 'brand', 'sales_price', 'image_urls__small']]
    recommendations_list = []

    for _, row in recommendations.iterrows():
        product_info = {
            "Product Name": row['product_name'],
            "Brand": row['brand'],
            "Price": row['sales_price'],
            "Image URL": row['image_urls__small']
        }
        recommendations_list.append(product_info)

    return recommendations_list

# Step 8: Test the Recommendation Function
product_index = 10  # Change to the index of the product you want to base recommendations on
top_n = 5
recommendations = recommend_products(product_index, top_n)

# Print Recommendations
for i, recommendation in enumerate(recommendations, start=1):
    print(f"Recommendation {i}:")
    for key, value in recommendation.items():
        print(f"  {key}: {value}")
    print("-" * 50)
```

```python
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from IPython.display import Image, display  # Import necessary functions to display images

# Step 1: Load the dataset
data_path = "C:/Users/sjvar/Downloads/fashion_products_data.ldjson"
data = pd.read_json(data_path, lines=True)

# Step 2: Check the dataset structure and ensure 'product_name' exists
if 'product_name' not in data.columns:
    raise ValueError("The column 'product_name' does not exist in the dataset. Please check the data structure.")

# Step 3: Prepare Product Names for Word2Vec
product_names = data['product_name'].fillna('').str.split()  # Tokenize product names

# Step 4: Train Word2Vec Model
word2vec_model = Word2Vec(
    sentences=product_names,  # Tokenized product names
    vector_size=100,          # Size of word embeddings
    window=5,                 # Context window size
    min_count=1,              # Minimum word frequency
    workers=4,                # Number of CPU threads
    seed=42                   # Random seed for reproducibility
)

# Step 5: Calculate TF-IDF for each word
corpus = [' '.join(words) for words in product_names]  # Re-join tokenized words into text for TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Step 6: Create Product Vectors by Averaging Word Vectors Weighted by TF-IDF
def get_weighted_product_vector(words, model, tfidf_vectorizer, tfidf_matrix):
    """Generate an averaged vector for a product name based on its words, weighted by TF-IDF."""
    tfidf_weights = dict(zip(tfidf_vectorizer.get_feature_names_out(), tfidf_matrix.toarray()[words.index]))  # Extract TF-IDF for words in product name
    weighted_vectors = []
    
    for word in words:
        if word in model.wv and word in tfidf_weights:
            word_vector = model.wv[word]
            weight = tfidf_weights[word]
            weighted_vectors.append(word_vector * weight)
    
    if weighted_vectors:
        return np.mean(weighted_vectors, axis=0)  # Average the weighted vectors
    else:
        return np.zeros(model.vector_size)  # Return a zero vector if no valid words

product_vectors = np.array([
    get_weighted_product_vector(words, word2vec_model, tfidf_vectorizer, tfidf_matrix) for words in product_names
])

# Step 7: Compute Similarity Matrix
cosine_sim = cosine_similarity(product_vectors)

# Step 8: Recommendation Function
def recommend_products(product_index, top_n=5):
    """
    Recommend products based on cosine similarity of product vectors.
    
    Args:
        product_index (int): The index of the product for which recommendations are to be made.
        top_n (int): Number of recommendations to return (default: 5).
    """
    if product_index >= len(data):
        print(f"Error: product_index {product_index} is out of range.")
        return

    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[product_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of top N similar products (excluding itself)
    top_indices = [i[0] for i in sim_scores[1:top_n + 1]]

    # Display Recommendations
    recommendations = data.iloc[top_indices][['uniq_id', 'product_name', 'brand', 'sales_price', 'image_urls__small']]
    recommendations_list = []

    for _, row in recommendations.iterrows():
        product_info = {
            "Product Name": row['product_name'],
            "Brand": row['brand'],
            "Price": row['sales_price'],
            "Image": row['image_urls__small']  # Store image URL for display
        }
        recommendations_list.append(product_info)

    return recommendations_list

# Step 9: Test the Recommendation Function
product_index = 10  # Change to the index of the product you want to base recommendations on
top_n = 5
recommendations = recommend_products(product_index, top_n)

# Print Recommendations and display images
for i, recommendation in enumerate(recommendations, start=1):
    print(f"Recommendation {i}:")
    for key, value in recommendation.items():
        if key == "Image":  # If the key is Image, display the image
            display(Image(url=value))  # Display the image inline in Jupyter Notebooks
        else:
            print(f"  {key}: {value}")
    print("-" * 50)
```
