```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=df['brand'], order=df['brand'].value_counts().index[:10])  # Top 10 brands
plt.title('Top 10 Brands by Product Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_ratings = df.groupby('brand')['rating'].mean().sort_values(ascending=False)[:10]  # Top 10 brands by rating
sns.barplot(x=avg_ratings, y=avg_ratings.index)
plt.title('Average Rating of Top 10 Brands')
plt.show()
```

```python
# Stock Availability by Brand
top_brands = df['brand'].value_counts().index[:10]
plt.figure(figsize=(12, 6))
sns.barplot(data=df[df['brand'].isin(top_brands)], x='brand', y='left_in_stock')
plt.title('Stock Availability by Brand')
plt.xlabel('Brand')
plt.ylabel('Stock Left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example of univariate analysis on sales price
sns.histplot(df['sales_price'].dropna(), kde=True)
plt.title('Sales Price Distribution')
plt.show()

# Univariate analysis on ratings
sns.histplot(df['rating'].dropna(), kde=True)
plt.title('Rating Distribution')
plt.show()
```

```python
# Bivariate Analysis
# Sales price vs rating
sns.scatterplot(data=df, x='sales_price', y='rating', alpha=0.5)
plt.title('Sales Price vs Rating')
plt.show()

# Sales price vs discount percentage
sns.scatterplot(data=df, x='sales_price', y='discount_percentage', alpha=0.5)
plt.title('Sales Price vs Discount Percentage')
plt.show()
```

```python
# Pair plot of selected numerical columns
sns.pairplot(df[['sales_price', 'rating', 'discount_percentage', 'no__of_reviews']].dropna())
plt.show()
```

```python
# Example for removing outliers in sales_price column
Q1 = df['sales_price'].quantile(0.25)
Q3 = df['sales_price'].quantile(0.75)
IQR = Q3 - Q1
df_filtered = df[(df['sales_price'] >= (Q1 - 1.5 * IQR)) & (df['sales_price'] <= (Q3 + 1.5 * IQR))]

print("Outliers removed dataset size:", df_filtered.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding for binary categorical column
df['amazon_prime_encoded'] = LabelEncoder().fit_transform(df['amazon_prime__y_or_n'])

# One-Hot Encoding for multi-category column like brand
df = pd.get_dummies(df, columns=['brand'], prefix='brand')
```

```python
# Heat Map for Numerical Data to Understand Collinearity
plt.figure(figsize=(10, 8))
sns.heatmap(df[['sales_price', 'rating', 'discount_percentage', 'no__of_reviews']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
