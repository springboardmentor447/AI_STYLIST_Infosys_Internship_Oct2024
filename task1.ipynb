```python
# 2. Categorical vs. Categorical
# Count plots for combinations of categorical pairs

# Select categorical columns (excluding any column that contains dict-like objects)
categorical_cols = data.select_dtypes(include=['object']).columns

# Function to check if a column contains dictionary-like values
def is_dict_column(col):
    try:
        # Check if any row in the column is a dictionary type
        return data[col].apply(lambda x: isinstance(x, dict)).any()
    except:
        return False

# Loop through the categorical columns and plot count plots for combinations of columns with fewer than 10 unique values
for i, col1 in enumerate(categorical_cols):
    if is_dict_column(col1):
        continue  # Skip columns with dictionary-like values
    for col2 in categorical_cols[i+1:]:
        if is_dict_column(col2):
            continue  # Skip columns with dictionary-like values
        # Check if both columns have fewer than 10 unique values
        if data[col1].nunique() < 10 and data[col2].nunique() < 10:
            plt.figure(figsize=(8, 6))
            sns.countplot(x=col1, hue=col2, data=data)
            plt.title(f'{col1} vs. {col2}')
            plt.xticks(rotation=45)
            plt.legend(title=col2)
            plt.show()
```

```python
# 3. Numerical vs. Categorical
# Boxplots and Violin plots
# Select numerical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Function to check if a column contains dictionary-like values
def is_dict_column(col):
    try:
        # Check if any row in the column is a dictionary type
        return data[col].apply(lambda x: isinstance(x, dict)).any()
    except:
        return False

# Loop through categorical columns and plot boxplots/violin plots
for cat_col in categorical_cols:
    if is_dict_column(cat_col):
        continue  # Skip columns with dictionary-like values
    
    # Limiting to a few unique categories for clear plots
    if data[cat_col].nunique() < 10:
        for num_col in numerical_cols:
            plt.figure(figsize=(10, 6))
            sns.boxplot(x=cat_col, y=num_col, data=data)
            plt.title(f'{num_col} vs. {cat_col} - Boxplot')
            plt.xticks(rotation=45)
            plt.show()

            # Violin plot for the same data to show distribution more clearly
            plt.figure(figsize=(10, 6))
            sns.violinplot(x=cat_col, y=num_col, data=data)
            plt.title(f'{num_col} vs. {cat_col} - Violin Plot')
            plt.xticks(rotation=45)
            plt.show()
```

```python
# Detect outliers in 'sales_price' using IQR method
Q1 = data['sales_price'].quantile(0.25)
Q3 = data['sales_price'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['sales_price'] < (Q1 - 1.5 * IQR)) | (data['sales_price'] > (Q3 + 1.5 * IQR))]

print(f"Outliers detected in 'sales_price': {len(outliers)}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot to visualize the distribution and outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['sales_price'])

# Highlight the outliers in red on the same plot
plt.scatter(outliers.index, outliers['sales_price'], color='red', label='Outliers')
plt.legend()

plt.title('Boxplot of Sales Price with Outliers Highlighted')
plt.show()
```
