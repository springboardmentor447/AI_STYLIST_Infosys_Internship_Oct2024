```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Define global parameters
img_width, img_height = 224, 224  # VGG16 input dimensions
save_path = "C:/Users/dhara/new_image_embeddings/"  # New folder to save embeddings
os.makedirs(save_path, exist_ok=True)  # Create folder if it doesn't exist

# Load VGG16 model (excluding the top layer for feature extraction)
model1 = VGG16(weights='imagenet', include_top=False)

# Define functions
def img_path(img_name):
    """Returns the full path to the image."""
    image_folder = r"C:/Users/dhara/.cache/kagglehub/datasets/paramaggarwal/fashion-product-images-small/versions/1/images/"
    return os.path.join(image_folder, img_name)

def model_predict(model, img_name):
    """Processes an image and extracts features using the provided model."""
    try:
        img = load_img(img_path(img_name), target_size=(img_width, img_height))
        x = img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        return model.predict(x).reshape(-1)
    except Exception as e:
        print(f"Error processing image {img_name}: {e}")
        return None

# Load dataset (styles.csv)
csv_path = r"C:/Users/dhara/.cache/kagglehub/datasets/paramaggarwal/fashion-product-images-small/versions/1/styles.csv"
df = pd.read_csv(csv_path, on_bad_lines="skip")

# Extract image column
df_copy1 = df.dropna(subset=["id"])  # Drop rows with NaN 'id'
df_copy1["image"] = df_copy1["id"].astype(str) + ".jpg"

# Process only the first 5000 images
df_subset = df_copy1.head(5000)

# Extract embeddings and save
all_embeddings = []
all_ids = []

for idx, row in df_subset.iterrows():
    img_name = row["image"]
    embedding = model_predict(model1, img_name)
    if embedding is not None:
        all_embeddings.append(embedding)
        all_ids.append(row["id"])
    print(f"Processed {len(all_embeddings)}/{len(df_subset)} images", end="\r")

# Save embeddings and IDs
np.save(os.path.join(save_path, "embeddings_5000.npy"), np.array(all_embeddings))
np.save(os.path.join(save_path, "ids_5000.npy"), np.array(all_ids))
print("\nEmbeddings and IDs saved for the first 5000 images.")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Paths to embeddings and CSV
embedding_path = "C:/Users/dhara/new_image_embeddings/embeddings_5000.npy"
ids_path = "C:/Users/dhara/new_image_embeddings/ids_5000.npy"
csv_file_path = r"C:\Users\dhara\.cache\kagglehub\datasets\paramaggarwal\fashion-product-images-small\versions\1\styles.csv"

# Step 1: Load data
embeddings = np.load(embedding_path)
ids = np.load(ids_path)
df = pd.read_csv(csv_file_path, on_bad_lines="skip")

# Step 2: Ensure IDs match embeddings
df_subset = df[df['id'].isin(ids)]

# Step 3: Normalize embeddings
from sklearn.preprocessing import normalize
embeddings_normalized = normalize(embeddings)

# Step 4: Create a recommendation function
def recommend(item_id, top_n=5):
    # Get the index of the item
    if item_id not in ids:
        return f"Item ID {item_id} not found in the processed dataset."
    
    item_idx = np.where(ids == item_id)[0][0]
    
    # Compute cosine similarities
    similarities = cosine_similarity([embeddings_normalized[item_idx]], embeddings_normalized)[0]
    
    # Get indices of top similar items (excluding itself)
    similar_indices = np.argsort(similarities)[::-1][1:top_n + 1]
    
    # Fetch similar items' IDs and scores
    similar_items = [(ids[i], similarities[i]) for i in similar_indices]
    
    return similar_items

# Step 5: Use the recommendation system
example_id = ids[0]  # Replace with an actual ID from your dataset
recommendations = recommend(example_id, top_n=5)

# Step 6: Display recommendations
print(f"Recommendations for Item ID {example_id}:")
for rec_id, score in recommendations:
    print(f"ID: {rec_id}, Similarity: {score:.4f}")

# Step 7: Join with the original DataFrame for details
recommended_items = df[df['id'].isin([rec[0] for rec in recommendations])]
print("\nDetails of recommended items:")
print(recommended_items)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize
from PIL import Image
import matplotlib.pyplot as plt
import os

# Paths to embeddings, IDs, CSV file, and images folder
embedding_path = "C:/Users/dhara/new_image_embeddings/embeddings_5000.npy"
ids_path = "C:/Users/dhara/new_image_embeddings/ids_5000.npy"
csv_file_path = r"C:\Users\dhara\.cache\kagglehub\datasets\paramaggarwal\fashion-product-images-small\versions\1\styles.csv"
image_folder = r"C:/Users/dhara/.cache/kagglehub/datasets/paramaggarwal/fashion-product-images-small/versions/1/images/"

# Step 1: Load data
embeddings = np.load(embedding_path)
ids = np.load(ids_path)
df = pd.read_csv(csv_file_path, on_bad_lines="skip")

# Step 2: Ensure IDs match embeddings
df_subset = df[df['id'].isin(ids)]

# Step 3: Normalize embeddings
embeddings_normalized = normalize(embeddings)

# Step 4: Create a recommendation function
def recommend_and_display(item_id, top_n=5):
    """
    Recommend top_n similar items based on cosine similarity and display images.

    Parameters:
    - item_id (int): ID of the item to find recommendations for.
    - top_n (int): Number of similar items to return.

    Displays:
    - The input image and the top recommended images.
    """
    if item_id not in ids:
        print(f"Item ID {item_id} not found in the processed dataset.")
        return
    
    # Get the index of the item in the embeddings
    item_idx = np.where(ids == item_id)[0][0]
    
    # Compute cosine similarities
    similarities = cosine_similarity([embeddings_normalized[item_idx]], embeddings_normalized)[0]
    
    # Get indices of top similar items (excluding itself)
    similar_indices = np.argsort(similarities)[::-1][1:top_n + 1]
    
    # Fetch similar items' IDs and scores
    similar_items = [(int(ids[i]), similarities[i]) for i in similar_indices]
    
    # Display the input image
    input_image_path = os.path.join(image_folder, f"{item_id}.jpg")
    input_image = Image.open(input_image_path)
    plt.figure(figsize=(15, 5))
    plt.subplot(1, top_n + 1, 1)
    plt.imshow(input_image)
    plt.axis('off')
    plt.title(f"Input ID: {item_id}")
    
    # Display recommended images
    for i, (rec_id, score) in enumerate(similar_items):
        recommended_image_path = os.path.join(image_folder, f"{rec_id}.jpg")
        try:
            recommended_image = Image.open(recommended_image_path)
            plt.subplot(1, top_n + 1, i + 2)
            plt.imshow(recommended_image)
            plt.axis('off')
            plt.title(f"ID: {rec_id}\nSim: {score:.2f}")
        except FileNotFoundError:
            print(f"Image for recommended ID {rec_id} not found.")
    
    plt.tight_layout()
    plt.show()

# Step 5: Call the recommendation function
item_id =  39386  # Replace with the actual item ID you want recommendations for
top_n = 5
recommend_and_display(item_id, top_n)
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Load the CSV file with image information
file_path = r"C:\Users\dhara\.cache\kagglehub\datasets\paramaggarwal\fashion-product-images-small\versions\1\styles.csv"
df = pd.read_csv(file_path, on_bad_lines='skip')

# Path to the image folder
image_folder = r"C:\Users\dhara\.cache\kagglehub\datasets\paramaggarwal\fashion-product-images-small\versions\1\images"

# Add the 'image' column with paths
df['image'] = df['id'].astype(str) + ".jpg"
df['image'] = df['image'].apply(lambda x: f"{image_folder}/{x}")

# Load the VGG16 model (without the top layer)
model2 = VGG16(weights="imagenet", include_top=False)

# Model prediction function
def model_predict(model, img_name):
    try:
        # Load and preprocess the image
        img = load_img(img_name, target_size=(224, 224))  # Resize to match VGG16 input size
        img_array = img_to_array(img)                     # Convert to array
        img_array = np.expand_dims(img_array, axis=0)     # Add batch dimension
        img_array = preprocess_input(img_array)           # Preprocess the input

        # Predict and return the flattened features
        features = model.predict(img_array)
        return features.flatten()

    except Exception as e:
        print(f"Error processing image {img_name}: {e}")
        return np.zeros((7 * 7 * 512,))  # Return a zero vector if image processing fails

# Train for the first 100 images
df_copy2 = df.head(100)  # Select only the first 100 rows
df_embedding2 = df_copy2['image'].apply(lambda x: model_predict(model2, x))
df_embedding2 = df_embedding2.apply(pd.Series)

# Print the shape of the embeddings and a few samples
print("Embeddings Shape:", df_embedding2.shape)
print("First 5 Embeddings:")
print(df_embedding2.head(5))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img

# Define image path
image_folder = r"C:/Users/dhara/.cache/kagglehub/datasets/paramaggarwal/fashion-product-images-small/versions/1/images/"

def find_similar(query_id, embeddings, ids, top_n=5):
   
    # Find the index of the query ID
    if query_id not in ids:
        print(f"ID {query_id} not found in the embeddings dataset.")
        return None
    
    query_idx = list(ids).index(query_id)
    query_embedding = embeddings[query_idx]
    
    # Compute cosine similarity
    similarities = cosine_similarity([query_embedding], embeddings).flatten()
    
    # Get top N similar products (excluding the queried product itself)
    similar_indices = similarities.argsort()[::-1][1:top_n+1]
    similar_ids = ids[similar_indices]
    similar_scores = similarities[similar_indices]
    
    return similar_ids, similar_scores

def display_images(query_id, similar_ids, similar_scores):
    """
    Displays the queried product and recommended products with their similarity scores.
    """
    # Display queried image
    query_image_path = os.path.join(image_folder, f"{query_id}.jpg")
    query_image = load_img(query_image_path)
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, len(similar_ids)+1, 1)
    plt.imshow(query_image)
    plt.title(f"Queried Product: {query_id}")
    plt.axis("off")
    
    # Display recommended images
    for i, (sim_id, score) in enumerate(zip(similar_ids, similar_scores), start=2):
        image_path = os.path.join(image_folder, f"{sim_id}.jpg")
        image = load_img(image_path)
        
        plt.subplot(1, len(similar_ids)+1, i)
        plt.imshow(image)
        plt.title(f"ID: {sim_id}\nScore: {score:.2f}")
        plt.axis("off")
    
    plt.show()

# Query example
query_id = 15970  # Replace with the ID you want to query
similar_ids, similar_scores = find_similar(query_id, embeddings, ids, top_n=5)

if similar_ids is not None:
    display_images(query_id, similar_ids, similar_scores)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
from PIL import Image
import os
import numpy as np

# Path to images
image_folder = r"C:/Users/dhara/.cache/kagglehub/datasets/paramaggarwal/fashion-product-images-small/versions/1/images/"

def show_image(img_id):
    """Displays the image given its ID."""
    img_path = os.path.join(image_folder, f"{img_id}.jpg")
    try:
        img = Image.open(img_path)
        plt.imshow(img)
        plt.axis("off")
        plt.show()
    except FileNotFoundError:
        print(f"Image for ID {img_id} not found!")

def recommend_product_manual(query_id, embeddings, ids, top_n=5):
    # Find the index of the query ID
    try:
        query_index = np.where(ids == query_id)[0][0]
    except IndexError:
        print("The queried ID is not in the dataset!")
        return

    # Get the embedding of the queried product
    query_embedding = embeddings[query_index]

    # Compute cosine similarity
    similarity_scores = cosine_similarity(
        query_embedding.reshape(1, -1), embeddings
    ).flatten()

    # Get top N similar indices (excluding the queried product itself)
    similar_indices = np.argsort(similarity_scores)[::-1][1 : top_n + 1]

    # Display the query image
    print(f"Queried Product (ID: {query_id}):")
    show_image(query_id)

    # Display recommended products
    print("Recommended Products:")
    for rank, idx in enumerate(similar_indices, start=1):
        recommended_id = ids[idx]
        score = similarity_scores[idx]
        print(f"Rank {rank}: ID = {recommended_id}, Similarity = {score:.4f}")
        show_image(recommended_id)

# Example Usage
def example_recommendation():
    # Load embeddings and IDs
    dense_embeddings = np.load("C:/Users/dhara/densenet_embeddings/densenet_embeddings_500.npy")
    dense_ids = np.load("C:/Users/dhara/densenet_embeddings/densenet_ids_500.npy")

    query_id = 39386  
    recommend_product_manual(query_id, dense_embeddings, dense_ids, top_n=5)


example_recommendation()
```
