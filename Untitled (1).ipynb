```python
import pandas as pd
import json
import re

# Load the data from the .ldjson file
file_path = 'C:/Users/admin/fashion_products_data.ldjson'  # Update this path if necessary
data = []

# Open and read data line by line
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        data.append(json.loads(line))

# Convert data into a DataFrame
df = pd.DataFrame(data)

# Columns to keep (product_name and rating)
columns_to_keep = ['product_name', 'rating']

# Check if the required columns exist in the dataset
missing_columns = [col for col in columns_to_keep if col not in df.columns]
if missing_columns:
    print(f"Missing columns in dataset: {missing_columns}")
else:
    # Keep only the product_name and rating columns
    df = df[columns_to_keep]

    # Define a set of common stopwords
    stop_words = set([
        'a', 'an', 'the', 'and', 'or', 'for', 'in', 'on', 'with', 'to', 'by', 'of',
        'at', 'is', 'it', 'this', 'that', 'these', 'those', 'as', 'are', 'was', 'were'
    ])

    # Basic lemmatizer for common English words
    def basic_lemmatizer(word):
        # Dictionary of common lemmas
        lemmas = {
            'feet': 'foot', 'geese': 'goose', 'children': 'child', 'men': 'man', 'women': 'woman',
            'mice': 'mouse', 'teeth': 'tooth', 'indices': 'index', 'data': 'datum'
        }
        # Singular and plural endings
        if word in lemmas:
            return lemmas[word]
        elif word.endswith('ies'):
            return word[:-3] + 'y'
        elif word.endswith('es') and word not in ['does', 'goes']:
            return word[:-2]
        elif len(word) > 1 and word.endswith('s') and word[-2] not in 'aeiou':  # Check length before indexing
            return word[:-1]
        return word

    # Porter Stemmer algorithm
    def porter_stemmer(word):
        # Porter stemming algorithm implementation
        suffixes = ['ing', 'ly', 'ed', 'ies', 'ive', 'es', 's', 'ment']
        if word.endswith('ing'):
            return word[:-3]
        for suffix in suffixes:
            if word.endswith(suffix):
                return word[:-len(suffix)]
        return word

    # Function to preprocess text by removing stopwords, lemmatizing, and applying basic stemming
    def preprocess_text(text):
        # Convert to lowercase and split into words
        words = re.findall(r'\b\w+\b', text.lower())
        # Remove stopwords, apply lemmatization and stemming
        processed_words = [
            porter_stemmer(basic_lemmatizer(word)) for word in words if word not in stop_words
        ]
        return ' '.join(processed_words)

    # Apply text preprocessing to the 'product_name' column
    df['product_name'] = df['product_name'].apply(preprocess_text)

    # Display the filtered data
    print("Product Names (with stopwords removed, lemmatized, and Porter stemming) and Ratings:")
    print(df)
```
